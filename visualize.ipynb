```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
```

```python
train_accuracy_avg = []
train_accuracy_std = []
train_f1_score_avg = []
train_f1_score_std = []
train_precision_avg = []
train_precision_std = []
train_recall_avg = []
train_recall_std = []
val_accuracy_avg = []
val_accuracy_std = []
val_f1_score_avg = []
val_f1_score_std = []
val_precision_avg = []
val_precision_std = []
val_recall_avg = []
val_recall_std = []
```

```python
dir = "./logs"
files = os.listdir(dir)
for file in files:
    if "train" in file:
        train = pd.read_csv(dir + "/" + file)
        train_accuracy_avg.append(train["accuracy"].mean())
        train_accuracy_std.append(train["accuracy"].std())
        train_f1_score_avg.append(train["f1_score"].mean())
        train_f1_score_std.append(train["f1_score"].std())
        train_precision_avg.append(train["precision"].mean())
        train_precision_std.append(train["precision"].std())
        train_recall_avg.append(train["recall"].mean())
        train_recall_std.append(train["recall"].std())
    elif "val" in file:
        val = pd.read_csv(dir + "/" + file)
        val_accuracy_avg.append(val["accuracy"].mean())
        val_accuracy_std.append(val["accuracy"].std())
        val_f1_score_avg.append(val["f1_score"].mean())
        val_f1_score_std.append(val["f1_score"].std())
        val_precision_avg.append(val["precision"].mean())
        val_precision_std.append(val["precision"].std())
        val_recall_avg.append(val["recall"].mean())
        val_recall_std.append(val["recall"].std())
    else:
        continue
```

```python
print(len(files), len(train_accuracy_avg), len(val_accuracy_avg), len(train_f1_score_avg), len(val_f1_score_avg), len(train_precision_avg), len(val_precision_avg), len(train_recall_avg), len(val_recall_avg))
```

```python
# plotting accuracy with respect to epochs with error bars
plt.figure()
# change the dimensions of the graph
plt.figure(figsize=(10, 5))
plt.errorbar(np.arange(1, len(train_accuracy_avg)+1), train_accuracy_avg, yerr=train_accuracy_std, fmt='-o', color='skyblue', ecolor='pink', elinewidth=1, capsize=0.5, markersize=2, linewidth=0.5, label="Train")
plt.title("Accuracy vs Epochs")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.show()
```
